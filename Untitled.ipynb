```python
# /////////////////// Object Blender ///////////////////////
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

img1=Image.open('datasets/car/testA/00004.jpg')
img2=Image.open('datasets/car/testA/00022.jpg')
    
    
    
# /////////////https://stackoverflow.com/questions/39382412/crop-center-portion-of-a-numpy-image/////////////////
def crop_center(img, cropx, cropy):
    y = img.shape[0]
    x = img.shape[1]
    
    startx = x // 2 - (cropx // 2)
    starty = y // 2 - (cropy // 2)
    return img[starty:starty + cropy, startx:startx + cropx , :]



def Embed_Images(im1 , mask , im2 , x , y , ratio):
#     a = crop_center(im1, 500 ,500)
#     print(a.shape)
#     imgplot = plt.imshow(a)

    #     ////////// This function get first images and it's binary mask and the seconf image to merge them togetehr ////////
    Target_V = im2.shape[0]
    Target_H = im2.shape[1]
        
    V = im1.shape[0]
    H = im1.shape[1]
    Aspect_ratio = H / V
    V = int(V * ratio)
    H = int(Aspect_ratio * V)
    
    newsize = (H, V)
    PIL_image = Image.fromarray(np.uint8(im1)).convert('RGB')
    im1 = PIL_image.resize(newsize, Image.ANTIALIAS)    
    
    im1 = np.asarray(im1)
    im1 = np.transpose(im1 , (2,0,1))
    im1 = im1.astype(int)
    
    padded_im1 = np.zeros((3 , Target_V , Target_H))
        
    top_pad = int((Target_V - V) * y )
    button_pad = int((Target_V - V) - top_pad)

    lef_pad = int((Target_H - H)  * x )
    right_pad = int((Target_H - H - lef_pad)) 
    
    padded_im1[0] = np.pad(im1[0],((top_pad , button_pad),(lef_pad , right_pad)), 'constant')
    padded_im1[1] = np.pad(im1[1],((top_pad , button_pad),(lef_pad , right_pad)), 'constant')
    padded_im1[2] = np.pad(im1[2],((top_pad , button_pad),(lef_pad , right_pad)), 'constant')
        
    mask = Image.fromarray(np.uint8(mask) , 'L')
    mask = mask.resize(newsize, Image.ANTIALIAS)  

    mask = np.asarray(mask)
    mask = mask.astype(int)
    
    padded_mask = np.zeros((Target_H , Target_V))  
    padded_mask = np.pad(mask,((top_pad , button_pad),(lef_pad , right_pad)), 'constant')
    padded_mask = np.where(padded_mask == 1 , 2, padded_mask)
    padded_mask = np.where(padded_mask == 0 , 1, padded_mask)
    padded_mask = np.where(padded_mask == 2 , 0, padded_mask)
    
    im2 = np.asarray(im2)
    im2 = np.transpose(im2 , (2,0,1))
    im2 = im2.astype(int)

    image = np.multiply(padded_mask,im2[:])
    image = np.add(padded_im1,image)
    image = np.transpose(image , (1,2,0))
    image = Image.fromarray(np.uint8(image)).convert('RGB')
    return (image)
    
def Transparent(img1):
    im = np.transpose(img1 , (2,0,1))
    im = im.astype(int)
    image = np.add(im[0,:,:],im[1,:,:])
    image = np.add(image,im[2,:,:])
    
    #     ///////////////////  Convert black points to 0 and white points to 1 ////////////////
    image_mask = np.where(image == 765, 0, image)
    image_mask = np.where(image_mask > 0 , 1, image_mask)
    #     ///////////////////// Coverting white background to black /////////////////
    image = np.multiply(image_mask,im[:])
    image = np.transpose(image , (1,2,0))
    return (image, image_mask )


image , mask = Transparent(img1)
x= 1 ; y= 0.5 ; ratio= 0.5
Embed_Image = Embed_Images(image , mask , np.asarray(img2) , x , y , ratio)

imgplot = plt.imshow(Embed_Image)
```
